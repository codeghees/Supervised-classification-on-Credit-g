```python
def removebyte(df):
    for col in list(df):
        print(col)
        if df[col].dtype != "float64":
            df[col] = df[col].apply(lambda x: x.decode('utf-8')) 
    return df
removebyte(df)
```

```python
def one_hot(df, cols):
    # https://stackoverflow.com/questions/37292872/how-can-i-one-hot-encode-in-python    
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode 
    @return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```python
X = model.apply(preprocessing.LabelEncoder().fit_transform)
Y =  X["class"]
```

```python
X = X.drop("class",axis =1 )
```

```python
X_values = X.values
print(len(X_values[0]))
```

```python
Y_values = Y.values
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X_values, Y_values, test_size = 0.3, random_state=0)
```

```python
clf = MultinomialNB()
clf.fit(xTrain, yTrain)
print("Accuracy is : ", clf.score(xTest,yTest)*100)
```

```python
clf = DecisionTreeClassifier(random_state=0,criterion="gini")
clf.fit(xTrain, yTrain)
print("Accuracy is : ", clf.score(xTest,yTest)*100)
```

```python
logiR = LogisticRegression(solver = 'lbfgs')
logiR.fit(xTrain, yTrain)
score = logiR.score(xTest, yTest)
print("The accuracy after logistic regression is = ", score*100)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
```

```python
#Source = https://gist.github.com/stevhliu/2dad7416cfb9eec107f3e5748cadbb2b#file-basicnn-py
m1 = Sequential()
m1.add(BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001, center=True, scale=True, 
                          beta_initializer='zeros', gamma_initializer='ones', moving_mean_initializer='zeros'))
m1.add(Dense(20, activation='relu', input_dim=8))
m1.add(Dropout(0.25))
m1.add(Dense(64, activation='relu'))
m1.add(Dropout(0.25))
m1.add(Dense(64, activation='relu'))
m1.add(Dropout(0.25))

m1.add(Dense(64, activation='relu'))
m1.add(Dropout(0.25))
m1.add(Dense(64, activation='relu'))
m1.add(Dropout(0.25))
m1.add(Dense(64, activation='relu'))
m1.add(Dropout(0.25))
m1.add(Dense(64, activation='relu'))
m1.add(Dense(2, activation='softmax'))
print(m1.summary)
# compile the model
m1.compile(loss='categorical_crossentropy',
           optimizer='adam',
           metrics=['accuracy'])

# fit the model
y_binary = np_utils.to_categorical(yTrain)
print(xTrain.shape)
print(y_binary.shape)
m1.fit(xTrain, y_binary,
       epochs=100,
       batch_size = 128)

# evaluate the model
m1.evaluate(xTrain, y_binary, batch_size=128)
```

```python
y_test = np_utils.to_categorical(yTest)

Results = m1.evaluate(xTest, y_test, batch_size=128)
print(Results[1]*100)
```
